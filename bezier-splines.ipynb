```python
import sympy as sp

t = sp.symbols('t', real=True)

def bezier_vector(points, t=t):
    """Return B(t) as a vector (Matrix) BÃ©zier curve."""
    n = len(points) - 1
    B = sp.zeros(points[0].rows, 1)
    for i, P in enumerate(points):
        B += sp.binomial(n, i) * (1 - t)**(n - i) * t**i * P
    return sp.simplify(B)
```
